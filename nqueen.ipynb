```python
def print_solution(board):
    for row in board:
        print(" ".join(str(x) for x in row))

def is_safe(board, row, col, N):
    # Check the current row on the left side
    if any(board[row][j] for j in range(col)):
        return False
    # Check upper and lower diagonals on the left side
    if any(board[i][col - (row - i)] for i in range(row) if col - (row - i) >= 0):
        return False
    if any(board[i][col - (i - row)] for i in range(row + 1, N) if col - (i - row) >= 0):
        return False
    return True

def solve_nqueens(board, col, N):
    if col >= N:
        return True
    for i in range(N):
        if is_safe(board, i, col, N):
            board[i][col] = 1
            if solve_nqueens(board, col + 1, N):
                return True
            board[i][col] = 0  # Backtrack
    return False

def n_queens(N):
    board = [[0] * N for _ in range(N)]
    if solve_nqueens(board, 0, N):
        print("Solution exists:")
        print_solution(board)
    else:
        print("Solution does not exist.")

if __name__ == "__main__":
    N = int(input("Enter the board size (n): "))
    n_queens(N)
```

```python
def is_safe(board, row, col, N):
    """
    Checks if it's safe to place a queen at the given position.

    Args:
        board: The current state of the chessboard.
        row: The row index of the position to check.
        col: The column index of the position to check.
        N: The size of the chessboard.

    Returns:
        True if it's safe, False otherwise.
    """

    # Check this row on left side
    for i in range(col):
        if board[row][i] == 1:
            return False

    # Check upper diagonal on left side
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    # Check lower diagonal on left side
    for i, j in zip(range(row, N, 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False

    return True

def solve_n_queens_util(board, col, N):
    """
    Recursively solves the N-Queens problem.

    Args:
        board: The current state of the chessboard.
        col: The current column to place a queen.
        N: The size of the chessboard.

    Returns:
        True if a solution is found, False otherwise.
    """

    if col >= N:
        return True

    for i in range(N):
        if is_safe(board, i, col, N):
            board[i][col] = 1

            if solve_n_queens_util(board, col + 1, N):
                return True

            board[i][col] = 0  # Backtrack

    return False

def solve_n_queens(N):
    """
    Solves the N-Queens problem.

    Args:
        N: The size of the chessboard.

    Returns:
        A list of lists representing the solution, or None if no solution exists.
    """

    board = [[0 for _ in range(N)] for _ in range(N)]

    # Place the first queen in the first row and first column
    board[0][0] = 1

    if not solve_n_queens_util(board, 1, N):
        return None

    return board

# Example usage:
N = 8
solution = solve_n_queens(N)

if solution:
    for row in solution:
        print(row)
else:
    print("No solution exists.")
```
